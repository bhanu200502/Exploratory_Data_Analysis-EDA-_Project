```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries Imported Successfully!")
```

```python
df = pd.read_csv("../dataset/Titanic-Dataset.csv")
df.head()
```

```python
print("Shape:", df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.drop('Cabin', axis=1, inplace=True)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.savefig("../outputs/age_distribution.png")
plt.show()
```

```python
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution")
plt.savefig("../outputs/gender_distribution.png")
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.savefig("../outputs/survival_count.png")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.savefig("../outputs/survival_gender.png")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.savefig("../outputs/survival_class.png")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(8,6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap")
plt.savefig("../outputs/heatmap.png")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.boxplot(y=df['Age'])
plt.subplot(1,2,2)
sns.boxplot(y=df['Fare'])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='Age',
    y='Fare',
    data=df
)
plt.title("Age vs Fare")
plt.savefig("../outputs/age_vs_fare.png")
plt.show()
```

Key Insights

Most passengers belonged to the 20–40 age group.
Male passengers were more numerous than female passengers.
Third-class passengers formed the largest group.
Female passengers had a significantly higher survival rate.
First-class passengers showed better survival chances.
Fare contained several high-value outliers.
Age had a weak correlation with survival.
Passenger class was one of the strongest factors influencing survival.
